```python
import matplotlib as plt
import seaborn as sb
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score

```

```python
bmw = pd.read_csv("bmw.csv")
bmw
```

```python
bmw.isnull().sum()
```

```python

bmw1 = pd.DataFrame({"year":bmw["year"],"transmission":bmw["transmission"]})
bmw1=bmw1.groupby(["year","transmission"]).size().reset_index(name="count")
bmw_wide = bmw1.pivot(index="year", columns="transmission", values="count").fillna(0)
bmw_wide = bmw_wide.reset_index()

print(bmw_wide)
```

```python
x = bmw_wide[["year"]]
y = bmw_wide[["Automatic",'Manual','Semi-Auto']]

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model =LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print("mean_squared _value",mean_squared_error(y_test,y_pred))
print("r_2",r2_score(y_test,y_pred))
```
